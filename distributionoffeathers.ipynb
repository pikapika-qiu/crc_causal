```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def plot_histograms(data, pdf_filename='histograms.pdf', plots_per_page=16):
    """
    Plots histograms of all numeric features in the DataFrame and saves them to a PDF file.

    Parameters:
    - data: pandas DataFrame containing the dataset.
    - pdf_filename: Name of the output PDF file.
    - plots_per_page: Number of plots per page (default is 16).
    """

    # Select only numeric columns
    numeric_data = data.select_dtypes(include=['number'])
    feature_names = numeric_data.columns.tolist()
    num_features = len(feature_names)

    # Calculate the number of pages needed
    num_pages = (num_features + plots_per_page - 1) // plots_per_page  # Ceiling division

    # Create a PdfPages object
    with PdfPages(pdf_filename) as pdf:
        for page in range(num_pages):
            # Create a new figure for each page
            fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(11.69, 8.27))  # A4 size in inches
            axes = axes.flatten()
            
            # For each subplot in the page
            for i in range(plots_per_page):
                feature_index = page * plots_per_page + i
                if feature_index < num_features:
                    feature_name = feature_names[feature_index]
                    ax = axes[i]
                    # Plot histogram
                    ax.hist(
                        numeric_data[feature_name].dropna(),  # Exclude NaN values
                        bins=30,
                        color='skyblue',
                        edgecolor='black'
                    )
                    ax.set_title(feature_name, fontsize=10)
                    ax.tick_params(axis='both', which='major', labelsize=8)
                else:
                    # Hide any unused subplots
                    axes[i].axis('off')
            plt.tight_layout()
            pdf.savefig(fig)
            plt.close(fig)

# Example usage:
# Assuming 'data' is your pandas DataFrame containing the dataset
# data = pd.read_csv('your_data.csv')
plot_histograms(data, pdf_filename='feature_histograms.pdf')
```
