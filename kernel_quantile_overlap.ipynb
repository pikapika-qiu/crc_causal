```python
import pandas as pd
from matplotlib_venn import venn2
import matplotlib.pyplot as plt
```

```python
data_kernel = pd.read_csv('/home/qiuaodon/Desktop/project_data_new/crc_50_bootstrap/crc_pseudo_gem_based_filtered_kernel_discrete_07202024.csv')
data_quantile = pd.read_csv('/home/qiuaodon/Desktop/project_data_new/crc_50_bootstrap/crc_pseudo_gem_based_filtered_quantile_discrete_07202024.csv')
```

```python

# Set 'Name' as the index for both dataframes
data_quantile.set_index('Name', inplace=True)
data_kernel.set_index('Name', inplace=True)

# For each value (-1, 0, 1), get the set of (row_name, column_name) where the cell equals that value
values = [-1, 0, 1]
features_dict = {}

for val in values:
    # For data_quantile
    mask_quantile = data_quantile == val  # Create a boolean mask
    positions_quantile = mask_quantile.stack()  # Stack to convert to a Series with MultiIndex
    positions_quantile = positions_quantile[positions_quantile].index  # Get indices where the mask is True
    set_quantile = set(positions_quantile)  # Convert to a set of (row_name, column_name)

    # For data_kernel
    mask_kernel = data_kernel == val
    positions_kernel = mask_kernel.stack()
    positions_kernel = positions_kernel[positions_kernel].index
    set_kernel = set(positions_kernel)

    # Store the sets in a dictionary
    features_dict[val] = {'Quantile': set_quantile, 'Kernel': set_kernel}

# Now, create Venn diagrams for each value
for val in values:
    quantile_set = features_dict[val]['Quantile']
    kernel_set = features_dict[val]['Kernel']
    
    plt.figure(figsize=(8, 8))
    venn = venn2([quantile_set, kernel_set], set_labels=('Quantile', 'Kernel'))
    plt.title(f'Venn Diagram for Value {val}')
    
    # Annotate the diagram with the counts
    if venn.get_label_by_id('10'):
        venn.get_label_by_id('10').set_text(len(quantile_set - kernel_set))
    if venn.get_label_by_id('01'):
        venn.get_label_by_id('01').set_text(len(kernel_set - quantile_set))
    if venn.get_label_by_id('11'):
        venn.get_label_by_id('11').set_text(len(quantile_set & kernel_set))
    
    plt.savefig(f'venn_{val}.png')
    plt.show()
```
